```python
from math import sin, pi
carrier_signal=[]
def ASK (fc,data):
    N=100 * fc
    tiv= 1/N#define el paso del tiempo teniendo en cuenta la frecuencia
    t=0
    s=[]#señal ask
    while(t< len(data)):
        carrier_signal.append(sin(2*pi*fc*t))
        s=s+[data[int(t)]*sin(2*pi*fc*t)]# se crea la señal ASK multiplicando la señal portadora y la digital
        t=t+tiv#paso del tiempo 
    return s
```

```python
import matplotlib.pylab as plt
import numpy as np
data=[1, 0, 0, 0, 0, 1, 1, 0, 1, 1, 1, 1, 0, 1, 0, 1, 0, 0, 0, 0]#señal digital dada en el ejercicio
fig = plt.figure(figsize=(30,15))
plt.title('Signal Carrier')
plt.xlabel('Time (t)')
plt.ylabel('carrier = sin(2*pi*fc*t)')
plt.plot(carrier_signal)#grafica la señal portadora
plt.show()
xs = np.repeat(range(len(data)), 2)
ys = np.repeat(data, 2)
xs = xs[1:]
ys = ys[:-1]
xs = np.append(xs,xs[-1] + 1)
ys = np.append(ys, ys[-1])
fig = plt.figure(figsize=(30,15))
plt.title('Digital Signal')
plt.xlabel('Time (t)')
plt.ylabel('S = Binary')
plt.plot(xs, ys)#grafica la señal binaria
plt.ylim(-0.5, 1.5)
plt.show()
fig = plt.figure(figsize=(30,15))
plt.title('ASK')
plt.xlabel('Time (t)')
plt.ylabel('ASK =carrier_signal * digital_signal')
plt.plot(ASK(2,data))#llama la función ASK que envia una frecuencia de 2 y la señal digital y la grafica
plt.show()


```
